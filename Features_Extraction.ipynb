```python
import pandas as pd
from tsfresh import extract_features
from tsfresh.feature_extraction import MinimalFCParameters
from sklearn.feature_selection import VarianceThreshold
```

```python
df = pd.read_csv("/content/drive/MyDrive/FitPulse_Milestone1/data/minute_level_data.csv")
df["timestamp"] = pd.to_datetime(df["timestamp"])
```

```python

df.columns
```

```python
user_ids = df["Id"].unique()[:14]
df = df[df["Id"].isin(user_ids)]
```

```python
df = df.groupby("Id").head(600)
```

```python
ts_data = df[["Id", "timestamp", "HeartRate"]].dropna()
ts_data.columns = ["id", "time", "value"]
```

```python
features = extract_features(
    ts_data,
    column_id="id",
    column_sort="time",
    default_fc_parameters=MinimalFCParameters(),
    disable_progressbar=False
)
```

```python
features.to_csv(
    "/content/drive/MyDrive/FitPulse_Milestone2/tsfresh_features.csv"
)
```
